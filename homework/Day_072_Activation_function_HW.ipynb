# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
# ReLU
x = np.arange(-10, 10, 0.01)
'''
def ReLU(x):
    if x >= 0:
        return x
    else:
        return 0
'''
y = list(map(lambda x: x if x>0 else 0, x))
plt.plot(x, y, 'r')
plt.grid()
plt.show()

dy = list(map(lambda x: 1 if x>0 else 0, x))
plt.plot(x, dy, 'b')
plt.grid()
plt.show()
```

```python
# leaky ReLU
x = np.arange(-10, 10, 0.01)
lambda_set = 0.1
y = list(map(lambda x: x if x>0 else x*lambda_set, x))
plt.plot(x, y, 'r')
plt.grid()
plt.show()

dy = list(map(lambda x: 1 if x>0 else 1*lambda_set, x))
plt.plot(x, dy, 'b')
plt.grid()
plt.show()
```
